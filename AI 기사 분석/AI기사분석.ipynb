```python
import requests
from bs4 import BeautifulSoup
import selenium
```

```python
#url 접속 확인
url = "https://www.joongang.co.kr/search/news?keyword=ai"
resp = requests.get(url)
bs = BeautifulSoup(resp.text, 'html.parser')

print(bs.prettify())
```

```python
#1chat gpt런칭 날짜 11월 30일 기준, 11월 28일 기사부터 크롤링
news_links = [] 
for i in range(50,0,-1):
    for j in ['인공지능','gpt']:
        url = f"https://www.joongang.co.kr/search/news?keyword={j}&page={i}"
        resp = requests.get(url)
        bs = BeautifulSoup(resp.text, 'html.parser')

        strong = bs.select_one('div.row')
        detail_urls = strong.select('section ul li div h2 a')
        for i in detail_urls:
            if i not in news_links:
                news_links.append(i.attrs['href'])

[print(a) for a in news_links]
print(len(news_links))
```

```python
#2. 크롤링한 url 저장
with open("C:\\Users\\ben81\\OneDrive\\바탕 화면\\joongang.txt", 'w', encoding='utf-8') as f:
    for i in news_links:
        f.write(i+'\n')
```
